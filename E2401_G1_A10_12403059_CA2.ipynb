### Create a story around year wise changes in the overall IPL performances. For that, get the records of year wise total run scored,total catches and total wickets taken, including all the player records.Comment on the year-wise variations in them and support your comments using proper visualizations.You are free to enrich the story on changes in year-wise IPL performances by using as many parameters from the original file as much you wish

```python
import pandas as pd
import matplotlib.pyplot as plt

data['Runs_Scored'] = pd.to_numeric(data['Runs_Scored'], errors='coerce')
data['Catches_Taken'] = pd.to_numeric(data['Catches_Taken'], errors='coerce')
data['Wickets_Taken'] = pd.to_numeric(data['Wickets_Taken'], errors='coerce')

data.fillna(0, inplace=True)
yearly_totals = data.groupby('Year')[['Runs_Scored', 'Catches_Taken', 'Wickets_Taken']].sum()
yearly_totals.plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title("Year-wise IPL Performance (Total Runs, Catches, Wickets)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Metrics")
plt.grid(axis='y', linestyle='--')
plt.tight_layout()

plt.show()
```
